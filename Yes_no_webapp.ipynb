```python
import transformers
import random

# Load the RoBERTa model and tokenizer
model = transformers.RobertaModel.from_pretrained('roberta-base')
tokenizer = transformers.RobertaTokenizer.from_pretrained('roberta-base')

def generate_response(question):
  # Encode the input text using the tokenizer
  input_ids = tokenizer.encode(question, return_tensors='pt').to('cpu')

  # Generate the final hidden state of the model
  hidden_state = model(input_ids).last_hidden_state

  # Calculate the logits for each response
  logits = hidden_state[:, 0, :]

  # Choose the response with the highest probability
  response_index = logits.argmax().item()
  responses = ["Yes", "No", "I'm not sure"]

  # Randomly select a response if there is a tie
  if logits[0][response_index] == logits.max().item():
    response_index = random.randint(0, 2)

  return responses[response_index]

print(generate_response("Are dogs pets?"))
```

```python
import gradio as gr
import transformers
import random
```

```python
inputs = gr.inputs.Textbox(placeholder="Ask a yes or no question")
outputs = gr.outputs.Textbox()
```

```python
def magic_8_ball(question):
  # Load the RoBERTa model and tokenizer
  model = transformers.RobertaModel.from_pretrained('roberta-base')
  tokenizer = transformers.RobertaTokenizer.from_pretrained('roberta-base')

  # Encode the input text using the tokenizer
  input_ids = tokenizer.encode(question, return_tensors='pt').to('cpu')

  # Generate the final hidden state of the model
  hidden_state = model(input_ids).last_hidden_state

  # Calculate the logits for each response
  logits = hidden_state[:, 0, :]

  # Choose the response with the highest probability
  response_index = logits.argmax().item()
  responses = ["Yes", "No", "I'm not sure"]

  # Randomly select a response if there is a tie
  if logits[0][response_index] == logits.max().item():
    response_index = random.randint(0, 2)

  return responses[response_index]
```

```python
app = gr.Interface(magic_8_ball, inputs, outputs)

app.launch(share=True)
```
